```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
```

```python
months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
sun_time = [2.8,3.9,5.0,6.2,7.4,8.3,8.9,7.8,6.3,4.8,2.8,2.4]
```

```python
plt.plot(months,sun_time)
plt.xlabel('Months')
plt.ylabel('Sun Time (Hours)')
plt.title('Monthly Sun Time in Toronto')
plt.show()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis,y_axis)
plt.show()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="maroon", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, marker="D", color="green", linewidth=2, label="Boston")
ax.set_xlabel("Months")
ax.set_ylabel("Fare ($)")
ax.set_title("PyBer Fare by Month")
ax.set_ylim(0,50)
ax.legend()
ax.grid()
```

```python
plt.scatter(y_axis,x_axis,color='red',label='Chicago',s=y_axis)
plt.gca().invert_yaxis()
plt.xlim(0,50)
plt.xlabel('Fares ($)')
plt.ylabel('Months')
plt.title('PyBer Fares by Month')
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)
```

```python
fig,ax=plt.subplots()
ax.scatter(y_axis, 
           x_axis, 
           s=[5*i for i in y_axis], 
           color="skyblue", 
           alpha=0.8,
           edgecolor='black',
           linewidth=2,
           label='Boston')
ax.legend()
ax.set_title('PyBer Fares by Month')
ax.set_xlabel('Fares ($)')
ax.set_ylabel('Months')
ax.set_xlim(0,50)
plt.gca().invert_yaxis()
```

```python
explode = (0,0,0.2,0,0,0,0.2,0,0,0,0,0)
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
```

```python
plt.subplots(figsize=(10,10))
plt.pie(y_axis,
        labels = x_axis,
        autopct='%1.1f%%',
        explode = explode,
        colors = colors)
plt.axis('equal')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,8))
ax.pie(y_axis,
        labels = x_axis,
        autopct='%1.1f%%',
        explode = explode,
        colors = colors,
        shadow = True,
        startangle = 270
      )
plt.gca().invert_yaxis()
```
